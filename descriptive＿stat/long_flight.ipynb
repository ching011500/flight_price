## Boxplot:平均價格_log

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ===== 1. 類別順序與資料排序 =====
# 出發時段
departure_order = ["凌晨班機", "早晨班機", "上午班機", "下午班機", "晚間班機"]
data['出發時段'] = pd.Categorical(data['出發時段'], categories=departure_order, ordered=True)
departure_data = data.sort_values(by='出發時段')

# 抵達時段
arrival_order = ["凌晨抵達", "早晨抵達", "上午抵達", "下午抵達", "晚間抵達"]
data['抵達時段'] = pd.Categorical(data['抵達時段'], categories=arrival_order, ordered=True)
arrival_data = data.sort_values(by='抵達時段')

# 航空聯盟對應字典
alliance_mapping = {
    1: '星空聯盟',
    2: '天合聯盟',
    3: '寰宇一家',
    4: '價值聯盟',
    0: '無聯盟'
}
data['航空聯盟'] = data['航空聯盟'].replace(alliance_mapping)

# 設定航空聯盟排序
alliance_order = ['星空聯盟', '天合聯盟', '寰宇一家', '價值聯盟', '無聯盟']
data['航空聯盟'] = pd.Categorical(data['航空聯盟'], categories=alliance_order, ordered=True)
alliance_data = data.sort_values(by='航空聯盟')

# 機型分類映射
category_mapping = {0: '窄體機', 1: '寬體機', 2: '超大型客機'}
data['機型分類'] = data['機型分類'].map(category_mapping)
# 機型分類排序
category_order = ['窄體機', '寬體機', '超大型客機']
data['機型分類'] = pd.Categorical(data['機型分類'], categories=category_order, ordered=True)
aircraft_data = data.sort_values(by='機型分類')

# ===== 2. 建立繪圖函式 =====
title_fontsize = 18
label_fontsize = 14

def plot_box(data, x_col, y_col, ax, title, x_label, y_label, color):
    """
    data  : DataFrame
    x_col : 分類欄位
    y_col : 數值欄位
    ax    : 子圖物件
    title : 圖表標題
    x_label, y_label : x, y 軸標籤
    color : boxplot 的填色
    """
    sns.boxplot(data=data, x=x_col, y=y_col, ax=ax, color=color)
    ax.set_title(title, fontsize=title_fontsize)
    ax.set_xlabel(x_label, fontsize=label_fontsize)
    ax.set_ylabel(y_label, fontsize=label_fontsize)
    ax.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)

# ===== 3. 建立子圖 (3x2) 並繪製圖表 =====
fig, axes = plt.subplots(3, 2, figsize=(20, 18))

# (1) 按出發時間
plot_box(departure_data, '出發時段', '平均價格_log', axes[0, 0],
         '按出發時間分類的平均價格（對數）', '出發時間', '平均價格（對數）', '#FF9999')

# (2) 按抵達時間
plot_box(arrival_data, '抵達時段', '平均價格_log', axes[0, 1],
         '按抵達時間分類的平均價格（對數）', '抵達時間', '平均價格（對數）', '#FFF2CC')

# (3) 按航空聯盟
plot_box(alliance_data, '航空聯盟', '平均價格_log', axes[1, 0],
         '按航空聯盟分類的平均價格（對數）', '航空聯盟', '平均價格（對數）', '#99CCFF')

# (4) 按機型分類
plot_box(aircraft_data, '機型分類', '平均價格_log', axes[1, 1],
         '按機型分類的平均價格（對數）', '機型分類', '平均價格（對數）', '#B3FFCC')

# (5) 按假期分類
plot_box(data, '假期', '平均價格_log', axes[2, 0],
         '按假期分類的平均價格（對數）', '假期', '平均價格（對數）', '#FFCC99')

# (6) 按出發機場代號
plot_box(data, '出發機場代號', '平均價格_log', axes[2, 1],
         '按出發機場代號分類的平均價格（對數）', '出發機場代號', '平均價格（對數）', '#CCCCFF')

plt.tight_layout()
plt.show()
```
