```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# è®€å–æ•¸æ“š
file_path = '/Users/yuchingchen/Documents/å°ˆé¡Œ/cleaned_data/short_flight_final.csv'
data = pd.read_csv(file_path)

# åˆ†é›¢ç¶“æ¿Ÿè‰™èˆ‡éç¶“æ¿Ÿè‰™çš„æ•¸æ“š
economy_class_data = data[data["è‰™ç­‰"] == "ç¶“æ¿Ÿè‰™"].copy()
non_economy_class_data = data[data["è‰™ç­‰"] != "ç¶“æ¿Ÿè‰™"].copy()

# One-hot encoding
categorical_cols = ['æ˜ŸæœŸ', 'å‡ºç™¼æ™‚æ®µ', 'å‡ºç™¼æ©Ÿå ´ä»£è™Ÿ', 'æŠµé”æ™‚æ®µ', 'æŠµé”æ©Ÿå ´ä»£è™Ÿ',
                    'èˆªç©ºå…¬å¸', 'èˆªç©ºè¯ç›Ÿ', 'æ©Ÿå‹åˆ†é¡', 'æ˜¯å¦éå¤œ', 'æ˜¯å¦ç‚ºå¹³æ—¥', 'æ©Ÿå‹', 'å‡æœŸ', 'Region']

economy_class_data = pd.get_dummies(economy_class_data, columns=categorical_cols, drop_first=True)
non_economy_class_data = pd.get_dummies(non_economy_class_data, columns=categorical_cols, drop_first=True)

# ç¢ºä¿å…©å€‹æ•¸æ“šé›†æ“æœ‰ç›¸åŒçš„ç‰¹å¾µæ¬„ä½
common_cols = list(set(economy_class_data.columns) & set(non_economy_class_data.columns))
economy_class_data = economy_class_data[common_cols]
non_economy_class_data = non_economy_class_data[common_cols]

# æ¨™æº–åŒ–æ•¸å€¼è®Šæ•¸
num_cols = ['é£›è¡Œæ™‚é–“_åˆ†é˜', 'ç¶“æ¿ŸæŒ‡æ¨™', 'æ©Ÿå ´æŒ‡æ¨™', 'competing_flights']
scaler = StandardScaler()
economy_class_data[num_cols] = scaler.fit_transform(economy_class_data[num_cols])
non_economy_class_data[num_cols] = scaler.transform(non_economy_class_data[num_cols])

# é¸å–å»ºæ¨¡ç‰¹å¾µ
target_keywords = ['å‡ºç™¼æ™‚æ®µ_', 'å‡ºç™¼æ©Ÿå ´ä»£è™Ÿ_', 'æŠµé”æ™‚æ®µ_', 'èˆªç©ºè¯ç›Ÿ_', 'æ©Ÿå‹åˆ†é¡_', 'æ˜¯å¦ç‚ºå¹³æ—¥_', 'å‡æœŸ_', 'Region_']
economy_class_data_dummy_cols = [col for col in economy_class_data.columns if any(keyword in col for keyword in target_keywords)]
non_economy_class_data_dummy_cols = [col for col in non_economy_class_data.columns if any(keyword in col for keyword in target_keywords)]
economy_class_data_feature_cols = economy_class_data_dummy_cols + ['é£›è¡Œæ™‚é–“_åˆ†é˜', 'ç¶“æ¿ŸæŒ‡æ¨™', 'æ©Ÿå ´æŒ‡æ¨™', 'competing_flights']
non_economy_class_data_feature_cols = non_economy_class_data_dummy_cols + ['é£›è¡Œæ™‚é–“_åˆ†é˜', 'ç¶“æ¿ŸæŒ‡æ¨™', 'æ©Ÿå ´æŒ‡æ¨™', 'competing_flights']

# å®šç¾©å‡½æ•¸ä¾†è¨“ç·´æ¨¡å‹ä¸¦é æ¸¬
def train_and_predict(X, y):
    # åˆ‡åˆ†è³‡æ–™ï¼ˆä¿ç•™æ¸¬è©¦é›†ç´¢å¼•ï¼‰
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    # è¨“ç·´éš¨æ©Ÿæ£®æ—
    rf_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    rf_model.fit(X_train, y_train)
    
    # é æ¸¬æ¸¬è©¦é›†
    y_pred_test = rf_model.predict(X_test)
    
    # è¨ˆç®— MSE & RÂ²
    mse = mean_squared_error(y_test, y_pred_test)
    r2 = r2_score(y_test, y_pred_test)
    print(f"ğŸ“Œ æ¸¬è©¦é›† MSE: {mse:.4f}")
    print(f"ğŸ“Œ æ¸¬è©¦é›† RÂ²: {r2:.4f}")
    
    return X_test.index, y_pred_test

# **ç¶“æ¿Ÿè‰™** é æ¸¬å¹³å‡åƒ¹æ ¼ï¼ˆåƒ…ç”¨æ¸¬è©¦é›†ï¼‰
economy_test_idx, economy_pred_log = train_and_predict(
    economy_class_data[economy_class_data_feature_cols], 
    economy_class_data["å¹³å‡åƒ¹æ ¼_log"]
)
# å°‡é æ¸¬çµæœ (è½‰å›åŸå§‹åƒ¹æ ¼) å¯«å…¥ç¶“æ¿Ÿè‰™çš„æ¸¬è©¦é›†éƒ¨åˆ†
economy_class_data = economy_class_data.loc[economy_test_idx].copy()
economy_class_data["é æ¸¬å¹³å‡åƒ¹æ ¼"] = np.exp(economy_pred_log)

# **éç¶“æ¿Ÿè‰™** é æ¸¬å¹³å‡åƒ¹æ ¼ï¼ˆåƒ…ç”¨æ¸¬è©¦é›†ï¼‰
non_economy_test_idx, non_economy_pred_log = train_and_predict(
    non_economy_class_data[non_economy_class_data_feature_cols], 
    non_economy_class_data["å¹³å‡åƒ¹æ ¼_log"]
)
# å°‡é æ¸¬çµæœ (è½‰å›åŸå§‹åƒ¹æ ¼) å¯«å…¥éç¶“æ¿Ÿè‰™çš„æ¸¬è©¦é›†éƒ¨åˆ†
non_economy_class_data = non_economy_class_data.loc[non_economy_test_idx].copy()
non_economy_class_data["é æ¸¬å¹³å‡åƒ¹æ ¼"] = np.exp(non_economy_pred_log)

# å„²å­˜é æ¸¬çµæœ
economy_class_data.to_csv('/Users/yuchingchen/Documents/å°ˆé¡Œ/model/data/short_economy_meanprice_predicted_rf.csv', index=False)
non_economy_class_data.to_csv('/Users/yuchingchen/Documents/å°ˆé¡Œ/model/data/short_non_economy_meanprice_predicted_rf.csv', index=False)

print("âœ… é æ¸¬å®Œæˆï¼Œå·²å„²å­˜ç‚º CSVï¼")
```

```python
import pandas as pd

# è®€å–é æ¸¬çµæœ
economy_file = "/Users/yuchingchen/Documents/å°ˆé¡Œ/model/data/short_economy_meanprice_predicted_rf.csv"
non_economy_file = "/Users/yuchingchen/Documents/å°ˆé¡Œ/model/data/short_non_economy_meanprice_predicted_rf.csv"

economy_class_data = pd.read_csv(economy_file)
non_economy_class_data = pd.read_csv(non_economy_file)

def recover_airport_code(df):
    """ æ‰¾å›åŸå§‹çš„ å‡ºç™¼æ©Ÿå ´ä»£è™Ÿ, æŠµé”æ©Ÿå ´ä»£è™Ÿ """
    
    # æ‰¾å‡ºæ‰€æœ‰å‡ºç™¼æ©Ÿå ´å’ŒæŠµé”æ©Ÿå ´çš„ One-Hot Encoding æ¬„ä½
    departure_cols = [col for col in df.columns if col.startswith("å‡ºç™¼æ©Ÿå ´ä»£è™Ÿ_")]
    arrival_cols = [col for col in df.columns if col.startswith("æŠµé”æ©Ÿå ´ä»£è™Ÿ_")]
    departure_time_cols = [col for col in df.columns if col.startswith("å‡ºç™¼æ™‚æ®µ_")]

    # é‚„åŸå‡ºç™¼æ©Ÿå ´ä»£è™Ÿ
    if departure_cols:
        df["å‡ºç™¼æ©Ÿå ´ä»£è™Ÿ"] = df[departure_cols].idxmax(axis=1).str.replace("å‡ºç™¼æ©Ÿå ´ä»£è™Ÿ_", "", regex=True)
    
    # é‚„åŸæŠµé”æ©Ÿå ´ä»£è™Ÿ
    if arrival_cols:
        df["æŠµé”æ©Ÿå ´ä»£è™Ÿ"] = df[arrival_cols].idxmax(axis=1).str.replace("æŠµé”æ©Ÿå ´ä»£è™Ÿ_", "", regex=True)

    # é‚„åŸå‡ºç™¼æ™‚æ®µ
    if departure_time_cols:
        df["å‡ºç™¼æ™‚æ®µ"] = df[departure_time_cols].idxmax(axis=1).str.replace("å‡ºç™¼æ™‚æ®µ_", "", regex=True)
    return df

economy_class_data = recover_airport_code(economy_class_data)
non_economy_class_data = recover_airport_code(non_economy_class_data)

# è¨ˆç®—æ•ˆèƒ½
economy_class_data["æ•ˆèƒ½"] = (economy_class_data["éš¨æ©Ÿè³¼è²·å¹³å‡åƒ¹æ ¼"] - economy_class_data["é æ¸¬å¹³å‡åƒ¹æ ¼"]) / economy_class_data["éš¨æ©Ÿè³¼è²·å¹³å‡åƒ¹æ ¼"]
non_economy_class_data["æ•ˆèƒ½"] = (non_economy_class_data["éš¨æ©Ÿè³¼è²·å¹³å‡åƒ¹æ ¼"] - non_economy_class_data["é æ¸¬å¹³å‡åƒ¹æ ¼"]) / non_economy_class_data["éš¨æ©Ÿè³¼è²·å¹³å‡åƒ¹æ ¼"]

# è¨ˆç®—æœ€ä½³æ•ˆèƒ½
economy_class_data["æœ€ä½³æ•ˆèƒ½"] = (economy_class_data["éš¨æ©Ÿè³¼è²·å¹³å‡åƒ¹æ ¼"] - economy_class_data["æœ€ä½åƒ¹æ ¼"]) / economy_class_data["éš¨æ©Ÿè³¼è²·å¹³å‡åƒ¹æ ¼"]
non_economy_class_data["æœ€ä½³æ•ˆèƒ½"] = (non_economy_class_data["éš¨æ©Ÿè³¼è²·å¹³å‡åƒ¹æ ¼"] - non_economy_class_data["æœ€ä½åƒ¹æ ¼"]) / non_economy_class_data["éš¨æ©Ÿè³¼è²·å¹³å‡åƒ¹æ ¼"]


# è¨ˆç®—æ¨™æº–åŒ–æ•ˆèƒ½
economy_class_data["æ¨™æº–åŒ–æ•ˆèƒ½"] = economy_class_data["æ•ˆèƒ½"] / economy_class_data["æœ€ä½³æ•ˆèƒ½"]
non_economy_class_data["æ¨™æº–åŒ–æ•ˆèƒ½"] = non_economy_class_data["æ•ˆèƒ½"] / non_economy_class_data["æœ€ä½³æ•ˆèƒ½"]

# ä¿ç•™ç‰¹å®šæ¬„ä½
selected_cols = ["å‡ºç™¼æ™‚é–“", "å‡ºç™¼æ™‚æ®µ", "å‡ºç™¼æ©Ÿå ´ä»£è™Ÿ", "æŠµé”æ©Ÿå ´ä»£è™Ÿ", "èˆªç­ä»£ç¢¼", "éš¨æ©Ÿè³¼è²·å¹³å‡åƒ¹æ ¼", "é æ¸¬å¹³å‡åƒ¹æ ¼", "å¹³å‡åƒ¹æ ¼", "æ•ˆèƒ½", "æœ€ä½³æ•ˆèƒ½", "æ¨™æº–åŒ–æ•ˆèƒ½"]
economy_class_data = economy_class_data[selected_cols]
non_economy_class_data = non_economy_class_data[selected_cols]

# å„²å­˜çµæœ
economy_output_file = "/Users/yuchingchen/Documents/å°ˆé¡Œ/model/data/short_economy_meanprice_performance_rf.csv"
non_economy_output_file = "/Users/yuchingchen/Documents/å°ˆé¡Œ/model/data/short_non_economy_meanprice_performance_rf.csv"

economy_class_data.to_csv(economy_output_file, index=False)
non_economy_class_data.to_csv(non_economy_output_file, index=False)

print("âœ… æ•ˆèƒ½è¨ˆç®—å®Œæˆï¼Œçµæœå·²å„²å­˜")
```

```python
# è®€å–é æ¸¬çµæœ
economy_file = "/Users/yuchingchen/Documents/å°ˆé¡Œ/model/data/short_economy_meanprice_performance_rf.csv"
non_economy_file = "/Users/yuchingchen/Documents/å°ˆé¡Œ/model/data/short_non_economy_meanprice_performance_rf.csv"

economy_class_data = pd.read_csv(economy_file)
non_economy_class_data = pd.read_csv(non_economy_file)

# éæ¿¾æ‰æœ€ä½³æ•ˆèƒ½ = 0 çš„èˆªç­
economy_class_data = economy_class_data[economy_class_data["æœ€ä½³æ•ˆèƒ½"] != 0]
non_economy_class_data = non_economy_class_data[non_economy_class_data["æœ€ä½³æ•ˆèƒ½"] != 0]

# è¨ˆç®—ç¶“æ¿Ÿè‰™çš„æ•ˆèƒ½
average_savings_economy = economy_class_data["æ•ˆèƒ½"].mean() * 100
best_possible_savings_economy = economy_class_data["æœ€ä½³æ•ˆèƒ½"].mean() * 100

# è¨ˆç®—éç¶“æ¿Ÿè‰™çš„æ•ˆèƒ½
average_savings_non_economy = non_economy_class_data["æ•ˆèƒ½"].mean() * 100
best_possible_savings_non_economy = non_economy_class_data["æœ€ä½³æ•ˆèƒ½"].mean() * 100

# æŒ‰å‡ºç™¼æ©Ÿå ´ & æŠµé”æ©Ÿå ´è¨ˆç®—å¹³å‡ç¯€çœæ¯”ä¾‹
savings_by_route = economy_class_data.groupby(["å‡ºç™¼æ©Ÿå ´ä»£è™Ÿ", "æŠµé”æ©Ÿå ´ä»£è™Ÿ"])["æ•ˆèƒ½"].mean() * 100

# æŒ‰å‡ºç™¼æ™‚æ®µä¾†çœ‹ç¯€çœæ¯”ä¾‹
savings_by_time = economy_class_data.groupby("å‡ºç™¼æ™‚æ®µ")["æ•ˆèƒ½"].mean() * 100

# çµ±è¨ˆæ¨™æº–åŒ–æ•ˆèƒ½çš„åˆ†ä½ˆ
efficiency_summary = economy_class_data["æ¨™æº–åŒ–æ•ˆèƒ½"].describe()

# è¨ˆç®—å¤§æ–¼ 1 çš„æ¯”ä¾‹ï¼ˆè¡¨ç¤ºé æ¸¬åƒ¹æ ¼ä½æ–¼å¸‚å ´æœ€ä½åƒ¹ï¼‰
efficiency_above_1 = (economy_class_data["æ¨™æº–åŒ–æ•ˆèƒ½"] > 1).mean() * 100

# è¨ˆç®—æ¥è¿‘ 1ï¼ˆÂ±0.05ï¼‰ çš„æ¯”ä¾‹ï¼ˆè¡¨ç¤ºæ¨¡å‹é æ¸¬æ¥è¿‘å¸‚å ´æœ€ä½åƒ¹ï¼‰
efficiency_near_1 = ((economy_class_data["æ¨™æº–åŒ–æ•ˆèƒ½"] >= 0.95) & (economy_class_data["æ¨™æº–åŒ–æ•ˆèƒ½"] <= 1.05)).mean() * 100

# è¨ˆç®—å°æ–¼ 1 çš„æ¯”ä¾‹ï¼ˆè¡¨ç¤ºæ¨¡å‹é‚„æ²’å®Œå…¨å­¸åˆ°å¸‚å ´æœ€ä½åƒ¹ï¼‰
efficiency_below_1 = (economy_class_data["æ¨™æº–åŒ–æ•ˆèƒ½"] < 1).mean() * 100

# å°å‡ºçµæœ
print(f"ğŸ“‰ ç¶“æ¿Ÿè‰™å¹³å‡ç¯€çœæ¯”ä¾‹: {average_savings_economy:.2f}%")
print(f"ğŸ“‰ éç¶“æ¿Ÿè‰™å¹³å‡ç¯€çœæ¯”ä¾‹: {average_savings_non_economy:.2f}%")
print(f"ğŸ¯ ç¶“æ¿Ÿè‰™æœ€ä½³ç¯€çœæ¯”ä¾‹: {best_possible_savings_economy:.2f}%")
print(f"ğŸ¯ éç¶“æ¿Ÿè‰™æœ€ä½³ç¯€çœæ¯”ä¾‹: {best_possible_savings_non_economy:.2f}%")

# å°å‡ºç¯€çœæœ€å¤šèˆ‡æœ€å°‘çš„èˆªç·š
print("ğŸ’° ç¯€çœæœ€å¤šçš„èˆªç·šï¼š")
print(savings_by_route.nlargest(5))
print("\nğŸ’¸ ç¯€çœæœ€å°‘çš„èˆªç·šï¼š")
print(savings_by_route.nsmallest(5))

# å°å‡ºä¸åŒå‡ºç™¼æ™‚æ®µçš„ç¯€çœæ¯”ä¾‹
print("\nğŸ•’ å‡ºç™¼æ™‚æ®µçš„ç¯€çœæ¯”ä¾‹ï¼š")
print(savings_by_time)

# å°å‡ºæ¨™æº–åŒ–æ•ˆèƒ½çš„çµ±è¨ˆ
print("\nğŸ“Š æ¨™æº–åŒ–æ•ˆèƒ½çš„çµ±è¨ˆï¼š")
print(efficiency_summary)
print(f"\nğŸ“ˆ æ¨™æº–åŒ–æ•ˆèƒ½ > 1 çš„æ¯”ä¾‹ï¼š{efficiency_above_1:.2f}%")
print(f"ğŸ“ˆ 0.95 <= æ¨™æº–åŒ–æ•ˆèƒ½ <= 1.05 çš„æ¯”ä¾‹ï¼š{efficiency_near_1:.2f}%")
print(f"ğŸ“ˆ æ¨™æº–åŒ–æ•ˆèƒ½ < 1 çš„æ¯”ä¾‹ï¼š{efficiency_below_1:.2f}%")
```
